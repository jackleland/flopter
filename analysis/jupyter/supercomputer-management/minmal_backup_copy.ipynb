```python
import pathlib as pth
import shutil
import os
import click
```

```python
data_folder = pth.Path('/home/jleland/data/external/spice')
os.chdir(data_folder)
```

```python
source_dir = pth.Path('marconi')
destination_dir = pth.Path('test')
```

```python
as_dirs = [folder for folder in source_dir.glob('*/*/*') if folder.is_dir()]
dirs = [folder for folder in source_dir.glob('*/*') if folder.is_dir() and folder not in as_dirs]

backup_dirs = [folder for folder in source_dir.glob('**/backup_*') if folder.is_dir()]
backup_dirs
```

```python
copy_groups = {}
confirm_group_fl = True

for bu_d in backup_dirs:
    run_name = bu_d.parent.name
    group_name = bu_d.parts[1]
    if confirm_group_fl: 
        if group_name not in copy_groups:
            copy_groups[group_name] = click.confirm(f'\nDo you want to copy the group folder "{group_name}"?', default=True)
        elif not copy_groups[group_name]:
            continue            
        
    print(f'\nCopying files for {run_name}')
    for bu_matfile in bu_d.glob(f'{run_name}.mat'):
        dest_file = destination_dir.joinpath(*bu_matfile.parts[1:])
        if not dest_file.exists():
            if not dest_file.parent.exists():
                print(f'\t Making path {dest_file.parent}')
                os.makedirs(dest_file.parent)
            print(f'\t Copying {bu_matfile} to {dest_file}')
            shutil.copy(bu_matfile, dest_file)
        else:
            print(f'\t Not copying {bu_matfile} as it already exists in {dest_file.parent}')
```
