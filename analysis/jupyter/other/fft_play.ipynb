```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 201)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.show()
```

```python
sp = np.fft.fft(y)
freq = np.fft.fftfreq(len(y), x[1] - x[0])

plt.figure()
plt.plot(freq, np.abs(sp))

plt.figure()
plt.plot(freq, np.angle(sp))
plt.show

print(freq[np.argmax(np.abs(sp))])
print(1/(2*np.pi))
```
