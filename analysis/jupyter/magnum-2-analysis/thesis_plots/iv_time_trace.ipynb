```python
fig, ax = plt.subplots(3, figsize=[8,8], sharex=True)

direc_ls = {
    'up': '--',
    'down': '-'
}
direc_m = {
    'up': 'x',
    'down': '+'
}
# probe_colours = {
#     'V': '#DDAA33',
#     'L': '#4477AA',
#     'S': '#EE6677',
#     'B': '#228833',
#     'R': '#CCBB44'
# }
probe_colours = {
    'V': '#DDAA33',
    'L': col.palettes['c'][3],
    'S': col.palettes['c'][2],
    'B': col.palettes['c'][1],
    'R': col.palettes['b'][2],
}

plot_ds = angle_scan_ds.sel(shot_number=36, sweep=slice(450,505)).mean('ts_radial_pos')

for sweep in plot_ds.sweep.values:
    for direction in ['up', 'down']:
        plot_sweep_ds = plot_ds.sel(sweep=sweep, direction=direction, probe='S')
    #     print(plot_sweep_ds.start_time.values)
        sweep_time = 0.01
        dt = sweep_time / (2 * plot_sweep_ds.time.size)
        if direction == 'up':
#             time = plot_sweep_ds.time.values / dt + plot_sweep_ds.start_time.values
            time = plot_sweep_ds.time.values + (2 * sweep * plot_sweep_ds.time.size)
            vline_ind = 0
        else:
#             time = np.flip(plot_sweep_ds.time.values) / dt + plot_sweep_ds.start_time.values
            time = np.flip(plot_sweep_ds.time.values) + ((2 * sweep + 1) * plot_sweep_ds.time.size)
            vline_ind = -1
        time = dt * time + plot_ds.start_time.min().values

        ax[0].plot(time, plot_sweep_ds.voltage.values, label=f'{sweep}:{direction}', 
                   ls=direc_ls[direction], color=probe_colours['V'])
        ax[1].plot(time, plot_sweep_ds.current.values, label=f'{sweep}:{direction}', 
                   ls=direc_ls[direction], color=probe_colours['S'])
        ax[2].plot(time, plot_ds.sel(probe='L', sweep=sweep, direction=direction).current.values, 
                   ls=direc_ls[direction], color=probe_colours['L'])
        for axis in ax:
            axis.axvline(x=time[vline_ind] - (0.5 * dt), **c.AX_LINE_DEFAULTS)
        
ax[0].set_ylabel('Probe Voltage [V]')
ax[1].set_ylabel('Probe Current [A]')
ax[2].set_ylabel('Probe Current [A]')
ax[2].set_xlabel('Shot Time [s]')
fig.tight_layout()
```
