# Combine all shots using open_mfdataset

```python
def preprocess_sel(ds):
    return ds.sel(sweep=slice(0,99))

def preprocess_average(ds):
    return ds.mean('direction').mean('sweep')
```

```python
for fn in analysed_dataset_fns:
    ds = xr.open_dataset(fn)
    print(fn, len(ds['sweep']))
```

```python
combined_ds = xr.open_mfdataset(analysed_dataset_fns, concat_dim='shot_number', preprocess=preprocess_average)
print(combined_ds)
```

```python
axial_profile = combined_ds.sel(probe='S').sel(shot_number=shot_numbers)

axial_profile = axial_profile.where(np.isfinite(axial_profile.current), drop=True)
print(axial_profile)
```
