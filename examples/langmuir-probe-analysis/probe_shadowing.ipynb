```python
thetas = np.linspace(0,10,100)
d = np.zeros_like(thetas)
h_coll = np.zeros_like(thetas)

for i, theta in enumerate(thetas):
    d[i], h_coll[i] = lpu.calc_probe_exposed_lengths(1, 0.081635, np.radians(theta), np.radians(10))

min_shad = thetas[(d!=0).argmin()]
print(thetas[(d!=0).argmin()])
    
plt.figure()
plt.plot(thetas, d, label=r'$d$')
plt.plot(thetas, h_coll, label=r'$h_{coll}$')
plt.axvline(x=min_shad, label='Min shadowing = {:.2g}'.format(min_shad), color='black', linestyle='--')
plt.legend()
plt.xlabel(r'Magentic field inclination angle ($\theta_{\perp}$) / degrees')
plt.ylabel('Length / mm')
```

```python
theta = 10
d_perps = np.linspace(0,1.0,100)
d1 = np.zeros_like(thetas)
h_coll1 = np.zeros_like(thetas)

for i, d_perp in enumerate(d_perps):
    d1[i], h_coll1[i] = lpu.calc_probe_exposed_lengths(1, d_perp, np.radians(theta), np.radians(10))

min_shad = d_perps[np.argmin(d1 + h_coll1)]
print(min_shad)
    
plt.figure()
plt.plot(d_perps, d1, label=r'$d$')
plt.plot(d_perps, h_coll1, label=r'$h_{coll}$')
plt.axvline(x=min_shad, label='Min shadowing = {:.2g}'.format(min_shad), color='black', linestyle='--')
plt.legend()
plt.title(r'Graph of exposed lengths for the L and M probes as a function of $d_{\perp}$')
plt.xlabel(r'Tile Drop ($d_{\perp}$) / mm')
plt.ylabel('Length / mm')
```

```python
theta = 10
d_perps = np.linspace(0,1.0,100)
d2 = np.zeros_like(thetas)
h_coll2 = np.zeros_like(thetas)

for i, d_perp in enumerate(d_perps):
    d2[i], h_coll2[i] = lpu.calc_probe_exposed_lengths(2, d_perp, np.radians(theta), np.radians(10))

min_shad = d_perps[np.argmin(d2 + h_coll2)]
print(min_shad)
    
plt.figure()
plt.plot(d_perps, d2, label=r'$d$')
plt.plot(d_perps, h_coll2, label=r'$h_{coll}$')
plt.axvline(x=min_shad, label='Min shadowing = {:.2g}'.format(min_shad), color='black', linestyle='--')
plt.legend()
plt.title(r'Graph of exposed lengths for the S probe as a function of $d_{\perp}$')
plt.xlabel(r'Tile Drop ($d_{\perp}$) / mm')
plt.ylabel('Length / mm')
```
